```python
from transformers import pipeline

generator  = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B", device_map="auto")

generator(
    "一個農場有雞跟兔，共有5個頭16隻腳，請問雞跟兔各有幾隻?",
    max_length=3000,
    num_return_sequences=1,
)

```
